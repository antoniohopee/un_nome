```python
#Esempio di funzione con più parametri
def galvanize_sentence(sentence):
    result = sentence + "!!!"
    return result

new_sentence = galvanize_sentence("Ciao")
print(new_sentence)
```

```python
#Esempio di funzione con più parametri
def super_galvanize_sentence(sentence, multiplier):
    result = sentence + "!!" * multiplier
    return result

new_sentence = super_galvanize_sentence("Ciao ragazzi", 3)
print(new_sentence)
```

```python
def square(number):
    return number ** 2

result = square(5)
print(result)
```

```python
# funzione che mi dice se un numero ha più di 5 cifre
def has_more_than_five_digits(number):
    if abs(number) > 99999:
        return True
    else:
        return False

random_number = 9999
result = has_more_than_five_digits(random_number)
print(result)
```

```python
# funzione che mi dice se un numero ha più di 1 cifra
def has_more_than_n_digits(number, digits_number):
    if abs(number) >= 10 ** (digits_number):
        return True
    else:
        return False

result = has_more_than_n_digits(12, 1)
print(result)
```
