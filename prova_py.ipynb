```python
# Esempio di variabile intera
a = 42
print("Il valore di a è:", a)
print("Tipo:", type(a))
```

```python
# Esempio di variabile float
x = 11
y = 3.14
print("x:", x, "Tipo:", type(x))
print("y:", y, "Tipo:", type(y))    
print("Somma:", x + y)
```

```python
# Liste annidate
nome = "Antonio"
saluto = "Ciao, " + nome + "!" 
print(saluto)
```

```python
# Esempio di variabile booleana
attivo = True
print("Stato attivo:", attivo)
print("Tipo di attivo:", type(attivo))
```

```python
# Esempio di variabile booleana
età = 20
maggiorenne = età >= 18
print("Età:", età)
print("Maggiorenne:", maggiorenne)
```

```python
# Operazioni sulle liste
frutti = ["mela", "banana", "ciliegia"]
print("Frutti:", frutti)
print("Tipo di frutti:", type(frutti))
```

```python
# Operazioni sulle liste
numeri = [1, 2, 3, 4, 5]
numeri.append(6,)
print("Numeri aggiornati:", numeri)
print("Lunghezza della lista numeri:", len(numeri))
print("Secondo elemento della lista numeri:", numeri[1])
print("Ultimo elemento della lista numeri:", numeri[-1])
```

```python
# Rimuovere il numero 3 dalla lista
numeri.remove(3)   
print("Numeri dopo la rimozione del 3:", numeri)
```

```python
# Verificare se "mela" è nella lista dei frutti
presente = "mela" in frutti
print("La mela è nella lista dei frutti?", presente)
```

```python
# Liste annidate
bibite = ["acqua", "soda", "succo"]
# Creare una lista della spesa che contiene frutti e bibite
spesa = frutti + bibite
print("Lista della spesa:", spesa)
print(spesa[0])
```

```python

import random

elemento = random.choice(spesa)
print(elemento)
```

```python
# Esempio di struttura condizionale
eta = 20

if eta >= 18:
    print("Sei maggiorenne.")
else:
    print("Sei minorenne.")
```

```python
if eta < 13:
    print("Sei un bambino.")
elif 13 <= eta < 18:
    print("Sei un adolescente.")
else:
    print("Sei un adulto.")
```
