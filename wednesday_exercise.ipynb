```python
#funzione che verifichi se una stringa è una targa italiana, cioè se è composta da 2 lettere, 3 numeri e 2 lettere, tutte maiuscole. La funzione deve restituire True o False a seconda del risultato.
def is_plate(candidate):
    candidate_init = candidate[0:2]
    init_cond = candidate_init.isalpha() and candidate_init.isupper() 
    candidate_middle = candidate[2:5]
    middle_cond = candidate_middle.isdigit()
    candidate_end = candidate[5:7]
    end_cond = candidate_end.isalpha() and candidate_end.isupper()
    lenght_cond = len(candidate) == 7
    if init_cond and middle_cond and end_cond and lenght_cond:
        return True
    else:
        return False
```

```python
new_variable = input("Enter a license plate: ")
result = is_plate(new_variable)
print(result)
```
